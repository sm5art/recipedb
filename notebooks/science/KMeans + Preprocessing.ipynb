```python
import sys
sys.path.insert(0, "/Users/artur/workspace/recipedb")
from recipedb.db import get_db
```

```python
from recipedb.science.preprocessing import Preprocessor
```

```python
preprocessor = Preprocessor()
```

```python
matched_recipes = preprocessor.matched_recipes
```

```python
db = get_db()
```

```python
Y_s = list(map(preprocessor.process_matched_recipe, matched_recipes))
```

```python
from sklearn.cluster import KMeans
import numpy as np
kmeans = KMeans(n_clusters=10, random_state=0).fit(np.array(Y_s))
```

```python
labels=kmeans.labels_
```

```python
where_is = 6
recipe_ids = list(map(lambda x: matched_recipes[x]['recipe_id'], list(np.where(labels==where_is)[0])))
```

```python
list(db.allrecipes.find({"_id":{"$in":recipe_ids}}))
```
